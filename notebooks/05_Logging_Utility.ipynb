```python
%sql
CREATE DATABASE IF NOT EXISTS mini_proj_logs;
```

```python
%sql
CREATE TABLE IF NOT EXISTS mini_proj_logs.etl_log (
    log_id BIGINT GENERATED ALWAYS AS IDENTITY,
    pipeline_name STRING,
    batch_id STRING,
    file_name STRING,
    layer STRING,
    status STRING,
    row_count BIGINT,
    message STRING,
    start_time TIMESTAMP,
    end_time TIMESTAMP
)
USING DELTA
LOCATION 's3a://mini-project-sathwik/logs/etl_log';
```

```python
def log_etl(pipeline_name, batch_id, file_name, layer, status, row_count, message):
    safe_message = message.replace("'", "''")
    spark.sql(f"""
        INSERT INTO mini_proj_logs.etl_log (
            pipeline_name, batch_id, file_name, layer,
            status, row_count, message, start_time, end_time
        )
        VALUES (
            '{pipeline_name}',
            '{batch_id}',
            '{file_name}',
            '{layer}',
            '{status}',
            {row_count},
            '{safe_message}',
            current_timestamp(),
            current_timestamp()
        )
    """)
```
